### a) Numerical Variables

```python
#This function allows us to create a boxplot and a histogram for any numerical variable that is given as input
#It also helps us to see the means and medians for the graphs
def hist_boxplot(df_cat,figsize=(15,7),bins=30):
    
    fig,(box_ax, hist_ax) = plt.subplots(nrows=2, 
                                        sharex=True, 
                                        gridspec_kw = {"height_ratios": (.25, .75)},
                                        figsize=figsize);

    sns.boxplot(df_cat,ax=box_ax, showmeans=True,color='skyblue');#For plotting the boxplot
    
    sns.distplot(df_cat, kde=False, ax=hist_ax, bins=bins); #Plotting the histogram
    hist_ax.axvline(df_cat.mean(), color='b', linestyle='--'); #Plotting the mean on the histogram with a blue dashed line
    hist_ax.axvline(df_cat.median(), color='r', linestyle='-'); #Plotting the median on the histogram with a red line
```

### Observations on Age

```python
#Calling the function hist_boxplot with the input variable 'Age' to get the boxplot and histogram as output
hist_boxplot(df['Age'], bins=32);
```

   ## 2) Bivariate Analysis

We shall plot a couple of bivariate graphs to understand the interactions the variables have with one another.  

### Correlation

```python
#We shall use a heat map to check the correlation between numerical variables

num_vals = ['Age', 'Education', 'Usage', 'Income', 'Miles']; #Array containing the names of numerical variables
fig,ax = plt.subplots(figsize=(15,10));
sns.heatmap(df[num_vals].corr(),ax=ax,annot=True, cmap='coolwarm', fmt='.1f');
```

#### Observations

* From the heat map, we can see most of the variables have positive correlation with each other
* A few of the correlations are on the weaker side with the correlation between miles and usage being the highest
* Miles shows high correlation with usage which suggests that as the number of times a customer uses the treadmill increases, so does the mileage for that customer
* The relationships between miles and age seems to be weak, the same can be said for the relationship between usage and age

### Bivariate Scatter Plots

```python
#This will plot the numerical variables against each other, creating various scatter plots on the variables
sns.pairplot(df[num_vals]);
```

#### Observations 

* We don't necessarily get a clear positive correlation at first glance to confirm what we saw in the heatmap between some of the variables but we can use a regression plot to help with that

```python
#We can create regression plots that shall draw a regression line--
#--through the scatter plot to help us visualize the positive correlation

plt.figure(figsize=(10,10))
sns.regplot(x='Usage',y='Miles', data=df); #Plotting a regression plot with Usage on the x-axis and Miles on the y-axis
```

```python
plt.figure(figsize=(10,10))
sns.regplot(x='Age',y='Income', data=df); #Plotting a regression plot with Age on the x-axis and Income on the y-axis
```

#### Observations

* These two graphs help us confirm what we saw in the heatmap which was the positive correlation Usage had with Miles, indicating that when usage increases so does mileage
* It also shows the positive correlation Age has with Income, suggesting income increases with age

# Conclusion and Recommendations

After all the analysis we are able to conclude that:

1. The majority of customers buying treadmills are between the ages of 20 to 35 with 25 year olds being the most frequent.

The TM195 might be the most efficient treadmill customers feel would satisfy their needs.

4. The model that is most used would be the TM798 which is used by customers with fitness levels ranging from 3-5. These customers are high to mid-income earners who are very active, having high mileages that tally with the high amount of times they use the treadmills in a week.

7. Partnered customers earn more in income at fitness level 3 but are less active than single customers although it isn't such a major difference in activity.

8. Males are generally more active than females except for at fitness level 5 where females output a higher turnout.
